Here we first check which columns have null values or any other string value.
The elimination of null value is simple as we can just do  df.dropna() in it.  but the indexed values  like 1[4] doesnot get filtered  from dropna().
The data is still dirty.

In the dataset, Peak and All Time Peak are having null and dirty values which makes the data dirty.
So in other to clean it we use the following step:

```python
df.isna().sum()
```

From Peak, we take out the unique values which are in the Peak column.

```python
df['Peak'].unique()
```

We extract the values using r'(\d+)'
r denotes raw string
\d means any digit 
+ means one or more digits
() means capture this part

```python
df['Peak']=df['Peak'].str.extract(r'(\d+)')
df=df.dropna(subset=['Peak'])
```

Now we get a fully clean data of Peak

# Finding the Hidden string of the columns 
some column name string may be hidden so we use repr(column_name) which will show its true form.

```python
for col in df.columns:
    print(repr(col))
```

```python
df.info()
```

```python
cols=['All Time Peak',
'Actual\xa0gross',
'Adjusted\xa0gross (in 2022 dollars)',
'Year(s)',

'Average gross']
df['Year(s)']=df['Year(s)'].str.extract(r'(\d{4})')
df['Year(s)']=pd.to_numeric(df['Year(s)'],errors='coerce')
for col in cols:
    df[col]=df[col].replace(r'[\$,]',"",regex=True)
    df[col]=pd.to_numeric(df[col],errors='coerce')
```
